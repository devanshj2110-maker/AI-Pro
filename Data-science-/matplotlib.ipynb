```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import pandas as pd 
```

```python
x=np.linspace(0,10,100)

sin=np.sin(x)
cos=np.cos(x)
plt.plot(x,sin,color='red',linestyle='-',linewidth=1,marker='x',mec='green',label='sin')
plt.plot(x,cos,color='purple',linestyle='-.',linewidth=2,marker='v',mec='red',markersize=4,label='cos')
plt.title("Sine-Cosine Graph")
plt.xlabel("X")
plt.ylabel("Sin(X)-Cos(X)")
plt.grid(True)
plt.legend(loc='lower right',fontsize=12)
# plt.xlim(0,20)
plt.ylim(3,-3)
plt.xticks([0,2,4,6,8,10],[0,'π/4','π/2','π','5π/4','3π/2'])
plt.show()
```

```python
csc=1/np.sin(x)
sec=1/np.cos(x)
plt.barh(x,sec)
plt.bar(x,csc)
plt.grid(True)
plt.title("Cosecant-Secant Graph")
plt.ylabel("Csc(X)-Sec(X)")
plt.xlabel("X")
plt.show()
```

```python
tan=np.tan(x)
cot=1/np.tan(x)
style.use('ggplot')
plt.plot(x,tan)
plt.plot(x,cot)
plt.title("Tangent-Cotangent Graph")
plt.xlabel("X")
plt.grid(True)
plt.ylabel("Tan(X)-Cot(X)")
plt.show()
```

```python
employee=['Alice', 'Bob', 'Carol', 'Dave', 'Eve', 'Frank']
salary=[70000, 80000, 120000, 90000, 65000, 75000]
years=[2, 5, 4, 3, 1, 2]
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
plt.pie(salary,labels=employee,autopct='%1.1f%%')
plt.subplot(2,2,2)
plt.hist(salary,bins=3,color='yellow',edgecolor='red')
plt.subplot(2,1,2)
plt.scatter(years,salary)
plt.grid()
# plt.tight_layout()
# plt.subplots_adjust(wspace=0.5)
plt.show()
```
